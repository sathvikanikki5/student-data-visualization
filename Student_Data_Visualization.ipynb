**📊 Student Data Visualization Project**

📂 Data Loading

```python
import pandas as pd

data = {
    "Name": ["A", "B", "C", "D", "E"],
    "Maths": [85, 70, 90, 60, 88],
    "Science": [78, 88, 92, 65, 79],
    "English": [92, 75, 89, 70, 85],
    "Total": [255, 233, 271, 195, 252]
}

df = pd.DataFrame(data)

df.head()
```

📈 Data Visualization

```python
import matplotlib.pyplot as plt

plt.bar(df["Name"], df["Total"])
plt.title("Total Marks")
plt.show()
```

```python
plt.figure()

plt.plot(df["Name"], df["Maths"], marker='o', label="Maths")
plt.plot(df["Name"], df["Science"], marker='o', label="Science")
plt.plot(df["Name"], df["English"], marker='o', label="English")

plt.title("Subject-wise Marks")
plt.legend()
plt.show()
```

```python
total_marks = df[["Maths", "Science", "English"]].sum()

plt.figure()
plt.pie(total_marks, labels=total_marks.index, autopct='%1.1f%%')
plt.title("Overall Subject Contribution")
plt.show()
```

```python
import seaborn as sns

plt.figure()
sns.heatmap(df[["Maths","Science","English"]], annot=True)
plt.title("Marks Heatmap")
plt.show()
```

🧠 Analysis

- Student C scored highest marks.
- Maths scores vary more compared to others.
- English scores are consistent.
